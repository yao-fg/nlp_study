#深度学习原理
概念知识：
    向量与标量概念：
    链式法则求导：
    激活函数：
    1.sigmoid
    2.tanh
    3.relu
    4.leaky relu
    5.softmax
    损失函数：
    1.MSE
    2.cross entropy
    梯度下降：
    反向传播：
    全连接层：
    神经网络的整体结构：输入层->若干隐藏层（全连接+激活）->输出层
    梯度下降三种变体：
        1.随机梯度下降:SGD
        2.批量梯度下降：BGD
        3.小批量梯度下降:Minibatch SGD
        --实践中最常用：Mini-batch SGD，批大小（Batch Size）通常取 32 / 64 / 128，多个批次组成一轮训练，遍历一次完整数据集 = 一个 Epoch
    训练流程总揽：前向传播->计算损失->反向传播->更新参数
    关键概念：epoch ,batch size
